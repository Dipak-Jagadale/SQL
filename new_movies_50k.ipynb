```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.imdb.com/search/title/?release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start=51&ref_=adv_nxt'
page=requests.get(url)
page.content
```

```python
soup=BeautifulSoup(page.content,'lxml')
soup
```

```python
all_movies=soup.find_all(class_="lister-item-content")
len(all_movies)
```

### Now let's find all movie data

```python
names, ratings, metas, votes, gross=[],[],[],[],[]
for f in all_movies:
    if f.find(class_="metascore"):
        names.append(f.find(class_="lister-item-header").get_text().replace('\n',''))
        ratings.append(f.find(class_="inline-block ratings-imdb-rating").get_text().replace('\n',''))
        metas.append(f.find(class_="metascore").get_text().strip())
        fvg=first.find(class_="sort-num_votes-visible").get_text().replace('\n','').split('|')
        if len(fvg)==2:
            votes.append(fvg[0].split(':')[1].replace(',',''))
            gross.append(fvg[1].split(':')[1])
        else:
            votes.append(fvg[0].split(':')[1].replace(',',''))
            gross.append(fvg[1].split(':')[1])
            
    else:
            names.append(f.find(class_="lister-item-header").get_text().replace('\n',''))
            ratings.append(f.find(class_="inline-block ratings-imdb-rating").get_text().replace('\n',''))
            metas.append(None)
            fvg=first.find(class_="sort-num_votes-visible").get_text().replace('\n','').split('|')
            votes.append(fvg[0].split(':')[1].replace(',',''))
            gross.append(None)
            
len(names), len(ratings), len(metas), len(votes), len(gross)            
            
```
